DATA VISUALIZATION

```python
df.hist() #hist() gives frequency distribution of the columns containing numeric data.
```

```python
#violinplot() function provides data distributions across entire categories.
sns.violinplot(x="Species",y="PetalWidthCm",data=df,split=True)
sns.violinplot(x="Species",y="SepalWidthCm",data=df,split=True)
```

```python
sns.heatmap(df.corr(),annot=True) #heatmap() results in a color-coded representation of 2D data.
```

```python
sns.pairplot(df,hue='PetalWidthCm') #pairplot gives data sets pairwise relationships.
sns.pairplot(df,hue='PetalLengthCm')
```

```python
#Using groupby() we can group dataset 
group=df.groupby(['Species','PetalLengthCm'])
x=group.size().unstack()
sns.heatmap(x,annot=True)
```

```python
#countplot on basis of Species and PetalWidthCm
sns.catplot(x="Species",hue="PetalWidthCm",kind="count",data=df)
```
