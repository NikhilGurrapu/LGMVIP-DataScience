```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv("globalterrorismdb_0718dist.csv",encoding="latin = 1")
df.head(10)
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType',
                   'target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df=df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df.head(10)
```

```python
df.corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
plt.xticks(fontsize=15,rotation=90)
plt.yticks(fontsize=15,rotation=0)
```

```python
df['Country'].value_counts().head(20)
```

```python
plt.figure(figsize=(50,20))
sns.countplot(df['Country'])
plt.title('Country vs Count of Attacks')
plt.xticks(rotation=70,fontsize=12)
plt.yticks(rotation=0,fontsize=12)
plt.xlabel('Country',fontsize=20)
plt.ylabel('No. of Attacks',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(50,20))
df['AttackType'].value_counts().plot.bar(color='orange')
plt.title('Most used Attacks',fontsize=25)
plt.xticks(rotation=70,fontsize=15)
plt.yticks(rotation=0,fontsize=15)
plt.xlabel('Attack type Names',fontsize=20)
plt.ylabel('No. of Attack Type',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(40,10))
df['AttackType'].value_counts().head(15).plot.pie()
plt.show()
```

```python
plt.figure(figsize=(40,10))
df['Country'].value_counts().tail(15).plot.pie()
plt.show()
```

```python
plt.figure(figsize=(40,10))
sns.countplot(df['Region'])
plt.title('Region vs Count of Attacks',fontsize=20)
plt.xticks(rotation=75,fontsize=15)
plt.yticks(rotation=0,fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Attacks',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(40,10))
df['Region'].value_counts().head(15).plot.pie()
plt.show()
```

```python
plt.figure(figsize=(40,10))
sns.barplot(x=df['Year'].value_counts()[:30].index,y=df['Year'].value_counts().values[:30])
plt.title('Number of Attacks per year',fontsize=20)
plt.xticks(rotation=75,fontsize=15)
plt.yticks(rotation=0,fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Attacks',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(30,5))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
df.Killed.plot(kind='line',color='red',label='No. of total fatalty death',linewidth=2,linestyle='--',alpha=1,figsize=(40,10),fontsize=20)
df.Wounded.plot(kind='line',color='green',label='No. of total non-fatal injury',linewidth=2,linestyle='solid',alpha=1,figsize=(40,10),fontsize=20)
plt.legend(loc='upper right',fontsize=25)
```

```python
df['Group'].value_counts().head()
```

```python
plt.figure(figsize=(40,20))
sns.countplot(df['Weapon_type'],order=df['Weapon_type'].value_counts().index)
plt.xticks(rotation=50,fontsize=15)
plt.yticks(rotation=0,fontsize=15)
plt.ylabel("No. of attacks",fontsize=20)
plt.xlabel("Weapons",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(40,20))
sns.barplot(x=df['Group'].value_counts()[1:20].index,y=df['Group'].value_counts()[1:20])
plt.xticks(rotation=50,fontsize=15)
plt.yticks(rotation=0,fontsize=15)
plt.ylabel("No. of Terrorist ",fontsize=20)
plt.xlabel("Groups",fontsize=20)
plt.title('Number of Terrorists in a Group',fontsize=20)
plt.show()
```

```python
pd.crosstab(df.Year,df.Region).plot(kind='line',figsize=(40,10))
plt.grid()
plt.title('Terrorist activities by Region in each Year',fontsize=25)
plt.ylabel("No. of Terrorist ",fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.yticks(rotation=0,fontsize=15)
plt.xticks([1970,1973,1976,1979,1982,1985,1988,1991,1994,1997,2000,2003,2006,2009,2012,2015,2017],rotation=90,fontsize=15)
plt.show()
```
